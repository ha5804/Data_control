```python
import numpy as np
import matplotlib.pyplot as plt
xaxis = np.array([2, 8])
yaxis = np.array([4, 9])

plt.plot(xaxis , yaxis)
plt.show()
```

```python
x = np.array([2, 12, 3, 9])
##default값은 y값이고, x범위 자동생성
plt.plot(x, marker = 'o', linestyle = "--")
plt.show()
```

```python
xaxis = ["A","B","C","D","E"]
yaxis = [5, 7, 8, 2, 6]
plt.bar(xaxis, yaxis, color = "green")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot Example")
plt.grid(True, axis = "y")
plt.show()
```

```python
np.random.seed(0)
#항상 같은 출력위해 randomseed 0으로 설정
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)

size = 1000 * np.random.rand(100)
#다른 크기의 scatter위해, x, y, colors, size를(0~1) random 100개 생성
#이때 size너무작아지는거 방지하기 위해 1000 곱해주기.

plt.scatter(x, y, c = colors, s = size, alpha = 0.5, cmap = "viridis")
#scatter에서는 color인자가 c이다.
plt.title("Scatter Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.colorbar()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (10, 10))
#subplots는 축 2개필요.
ax1.plot(x, y1, color = "blue", label = "Sine Wave")
ax1.set_title("Sine Wave")
ax1.set_xlabel("X Axis")
ax1.set_ylabel("Y Axis")
#ax는 plt.title -> ax.set_title
ax1.legend()
ax1.grid(True)

ax2.plot(x, y1, color = "red", label = "Cosine Wave")
ax2.set_title("Cosine Wave")
ax2.set_xlabel("X Axis")
ax2.set_ylabel("Y Axis")
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
data = {"Date": ["2024-10-01", "2024-10-01", "2024-10-02",
                 "2024-10-02", "2024-10-03", "2024-10-03"],
                 "Product" : ["A","B","A","B","A","B"],
                 "Sales" : [150, 200, 180, 220, 170, 210]}
df = pd.DataFrame(data)

sales = df.groupby("Product")["Sales"].sum()
a, b = sales

sales_a = df.groupby("Product")["Sales"].sum()["A"]
sales_b = df.groupby("Product")["Sales"].sum()["B"]


mean_a = df.groupby("Product")["Sales"].mean()["A"]
mean_b = df.groupby("Product")["Sales"].mean()["B"]
print(sales_a,sales_b,f"{mean_a:.2f}",mean_b)

A = df[df["Product"] == "A"]
B = df[df["Product"] == "B"]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (10,10))
ax1.plot(A["Date"], A["Sales"], color = "orange", marker = 'o', label = "A")
ax1.plot(B["Date"], B["Sales"], color = "blue", marker = 'o', label = "B")
ax1.set_title("Daily Sales Trend by Product")
ax1.legend(title = "Product")

ax2.bar(["A", "B"], [sales_a, sales_b])
ax2.set_title("Total Sales by Product")
plt.tight_layout()
plt.show()



            
```
