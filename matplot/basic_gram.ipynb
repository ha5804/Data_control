# 1.Structure

#### (1) Figure and Axes

- figure is the top-level container.

it includes everthing visualized in a plot including one or more Axes.


- Axes contain most of the elements in a plot

Axis, Tick, Line2D, Text...

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (8,8))
```

#### (2) plot

plot draw line in arange.



```python
import numpy as np
xaxis = np.array([2,8])
yaxis = np.array([4,9])
plt.plot(xaxis, yaxis)
plt.show()
```

#### (3) customize marker & linestyle

if you not set (x,y), matplot is automatically set xrange.

so if you just one argument, it encoded y.

```python
xaxis = np.array([2, 12, 3, 9])
plt.plot(xaxis, marker = "o", linestyle = "--")
plt.show()
```

# 2. Ex line plot

```python
x = np.linspace(0,10,100)
#np.linspace(start, stop, num divided, end_point = True)
y = np.sin(x)

plt.figure(figsize = (10,6))
plt.plot(x, y, label = "Sine Wave", color = "blue", linewidth = 2)
plt.title("Line Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.grid(True)
plt.show()
```

# 3.Ex Bar Plot

```python
categories = ["A","B","C","D","E"]
values = [5,7,8,2,6]

plt.figure(figsize = (10,6))
plt.bar(categories, values, color = "Green")
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(axis = "y")
plt.show()
```

# 3.Ex Scatter Plot

```python
np.random.seed(0)
#always same result each run time. (fixed random seed)

x = np.random.rand(100)
y = np.random.rand(100)
#(0 ~ 1) random number 100

colors = np.random.rand(100)
#(0 ~ 1) random colors 100

size = 1000 * np.random.rand(100)
#(0 ~ 1) random size 100

plt.figure(figsize = (10,6))
plt.scatter(x, y, c= colors, s=size, alpha =0.5 ,cmap = "viridis")
plt.title("Scatter Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.colorbar()
plt.show()
```

# 4.Sub Plot

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (10,10))
#plt.figure -> fig, (ax1, ax2) -> plt.subplots(row, col, figsize =())
ax1.plot(x, y1, label = "Sine Wave", color = "blue")
ax1.set_title("Sine Wave")
ax1.set_xlabel("X Axis")
ax1.set_ylabel("Y Axis")
ax1.legend()
ax1.grid(True)

ax2.plot(x, y2, label= "Cosine Wave", color = "red")
ax2.set_title("Cosine Wave")
ax2.set_xlabel("X Axis")
ax2.set_ylabel("Y Axis")
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
