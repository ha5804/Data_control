```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Quarters = ["Q1", "Q2", "Q3", "Q4"]
sold = [150, 200, 250, 300]

plt.figure(figsize = (8,8))
plt.xlabel("Quater")
plt.ylabel("Products sold")
plt.bar(Quarters, sold)
plt.grid(True)
plt.title("Quarterly Product Sales")
```
